```python
import pandas as pd
import seaborn as sns
import csv
import matplotlib.pyplot as plt
```

```python
def plot_genre_sales(file):
    # Read the CSV file into a DataFrame
    df = pd.read_csv(file)

    # Filter the DataFrame for action and adventure games
    action_adventure_df = df[(df['genre'] == 'Action') | (df['genre'] == 'Adventure')]

    # Group by Genre and sum the Global_Sales
    genre_sales = action_adventure_df.groupby('genre')['na_sales'].sum()

    genre_sales.plot(kind='bar', color=['blue', 'green'])
    plt.title('North America Sales of Action and Adventure Games')
    plt.xlabel('Genre')
    plt.ylabel('North America Sales (in millions)')
    plt.show()

file = 'vgchartz-2024.csv'
plot_genre_sales(file)
```

```python
def plot_individual_game_sales(file):
    df = pd.read_csv(file)
    action_adventure_df = df[(df['genre'] == 'Action') | (df['genre'] == 'Adventure')]
    
    # Sort by North America Sales in descending order and select the top 20
    top_20_na_sales = action_adventure_df.sort_values(by='na_sales', ascending=False).head(20)

# Plot individual games based on North America sales
    plt.figure(figsize=(12, 6))
    plt.bar(top_20_na_sales['title'], top_20_na_sales['na_sales'], color='skyblue')
    plt.title('Top 20 Overall North America Sales of Action and Adventure Games')
    plt.xlabel('Game Name')
    plt.ylabel('North America Sales (in millions)')
    plt.xticks(rotation=90, ha='right')
    plt.tight_layout()
    plt.show()

file = 'vgchartz-2024.csv'
plot_individual_game_sales(file)
```

```python
file_path = 'vgchartz-2024.csv'
df = pd.read_csv(file_path)

developers = df['developer'][:20]
sales = df['na_sales'][:20]

plt.figure(figsize=(12, 6))
plt.bar(developers, sales, color='blue')
plt.xlabel('Developers')
plt.ylabel('NA Sales (in millions)')
plt.title('Top Action and Adventure Video Game Developers')
plt.xticks(rotation=90, ha='right')
plt.tight_layout()

plt.show()
```
